```python
# import libraries
import numpy as np # 維度陣列與矩陣運算library
import matplotlib.pyplot as plt # 資料視覺化library
```

```python
# define a function y = x^2 
def y_x_square(x):
  return np.square(x)
```

```python
# 視覺化 y_x_square 這個 function 的樣子

x = np.linspace(-10, 10, 100) # x 軸設定 -10 到 10 細分為 100個點
y = y_x_square(x)
plt.plot(x,y)
plt.show()
```

```python
# define first derivative of y_x_square => dy/dx=2*x 
def de_func(x): 
  return 2 * x
```

```python
# define gradient descent function as GD_func 
# there are four input and one output 
def GD_func(x_start, de_func, epochs, learn_rate):  
  
  ''' 
  1. choose a starting point (x_start)
  2. calculate gradient of the point (de_func)
  3. make a scaled step (learn_rate) in the opposite direction

  ''' 
  x = x_start 
  x_new = np.zeros(epochs+1) 
  x_new[0]=x   
     
  for i in range(epochs):         
      dx = de_func(x)      
      scale = -1 * learn_rate * dx        
      x += scale # x = x + scale        
      x_new[i+1] = x    
  return x_new

# Main code
x_start = 8    
epochs = 15 
lr = 0.1   
x_new = GD_func(x_start, de_func, epochs, learn_rate=lr) 
print (x_new)
```

```python
x = np.linspace(-10, 10, 100)

plt.plot(x, y_x_square(x), c='b')
plt.plot(x_new, y_x_square(x_new), c='r', label='Learning rate={}'.format(lr))    
plt.scatter(x_new, y_x_square(x_new), c='r',) 

plt.legend()
plt.show()
```
