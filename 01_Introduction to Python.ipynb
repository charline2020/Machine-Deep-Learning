**List comprehensions**

```python
nums = [0,1,2,3,'4']
squares = []
for x in nums:
  squares.append(int(x)**2)
```

```python
print(squares)
```

```python
print([int(x)**2 for x in nums])
```

```python
even_squares=[int(x)**2 for x in nums if int(x)%2==0]
print(even_squares)
```

**User input**

```python
catNames = []
flag = True
while flag:
  print('Enter the name of cat ' + str(len(catNames)+ 1) + ' (Or enter nothing to stop.):')
  name = input()
  if name == '': # no imput
        flag = False
  catNames = catNames + [name] # list concatenation # 串列連續

print('The cat names are:')
for name in catNames:
  print('  ' + name)
```

**User-defined functions**

```python
def sample1(*args):
  for a in args:
    print(a)

sample1(1,2,3,4,5)
```

```python
def sums(x, y):
  return x + y

sums(2,6)
```

```python
def sample2(**kwargs):
  for a in kwargs:
    print(a)
    print(kwargs[a])

sample2(name='John', age=27)
```

```python
def mix(*args, **kwargs):
  get_args = args
  get_kwargs = kwargs

  for key in kwargs:
    print("key:", key)
    print("value", kwargs[key])
  
  return get_args, get_kwargs

get_args, get_kwargs = mix(1, 2, 3, 4, apple=45, ball=100, cup=300)
```

```python
get_args, get_kwargs = mix(1, 2, 3, 4, apple=45, ball=100, cup=300)

print("get_args return:", get_args)
print("get_kwargs return:",get_kwargs)
```

**Plot**

```python
import matplotlib.pyplot as plt
year = [1950, 1970, 1990, 2010]
population = [2.519, 3.692, 5.263, 6.972]
plt.plot(year, population)
plt.show()
```

```python
plt.plot(year, population, color='r', marker = 'o') # change line color, data point marker
plt.title("Year Population Plot") # image title
plt.ylabel("Population", fontname='Arial', fontsize = 20) # y label, label font style and size
plt.xlabel("Year") # x label


plt.show()
```

```python
plt.scatter(year, population)
plt.show() 
```

```python
plt.bar(year, population )
plt.show()
```
