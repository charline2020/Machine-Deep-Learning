**Managing categorical data**

```python
import numpy as np
import pandas as pd
country=['Taiwan','Australia','Ireland','Australia','Ireland','Taiwan']
age=[25,30,45,35,22,36]
salary=[20000,32000,59000,60000,43000,52000]

data = pd.DataFrame({'Country':country,'Age':age,'Salary':salary})
print(data)
```

**LabelEncoder**

```python
from sklearn.preprocessing import LabelEncoder # import library
labelencoder = LabelEncoder() 
data_c = data['Country'] # 要編碼的類別是Country
data_c_encode = labelencoder.fit_transform(data_c) # 編碼過後的結果會回傳給 data_c_encode
print(labelencoder.classes_) # encoder 之後的類別順序 0 1 2 分別為 'Australia' 'Ireland' 'Taiwan'
print(data_c_encode)
```

```python
data['Country'] = labelencoder.fit_transform(data_c) # 編碼過後的結果會回傳給 data中 Country這一列
print(data)
```

**OneHotEncoder**

```python
from sklearn.preprocessing import LabelBinarizer
encoder = LabelBinarizer()
data_c_1hot = encoder.fit_transform(data['Country'])
print(data_c_1hot) # onehotencoder 結果
```

```python
data_dum = pd.get_dummies(data) # get_dummies 一樣可以將類別資料轉換成數值形式的資料
pd.DataFrame(data_dum)
```

**Data scaling and normalization**

```python
from sklearn.preprocessing import MinMaxScaler
data = np.array([[1., -1., 2.], 
                [2., 0., 0.], 
                [0., 1., -1.]])  
min_max_scaler = MinMaxScaler()   
new_data = min_max_scaler.fit_transform(data) 
print(new_data)   
```

```python
from sklearn.preprocessing import StandardScaler
data = np.array([[1., -1., 2.], 
                 [2., 0., 0.], 
                 [0., 1., -1.]])  
zscore = StandardScaler()   
new_data = zscore.fit_transform(data) 
print(new_data) 
```
